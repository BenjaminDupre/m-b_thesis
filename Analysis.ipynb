#### Meta-Data Analysis

There were a total of 23 participans who performed both of the full tasks. There were considerably more women (17) than men (5) in the sample and the age was consistenly 25 years old. <br> ($\mu=25.1 , \sigma=6.3$)


```python
# Loading Libraries
import numpy as np  # not sure for what
import scipy as sp
import pandas as pd 
import seaborn as sb
import os 
import matplotlib.pyplot as plt
# Loading Data
notebook_path = os.path.abspath("Analysis.ipynb")
metadat  = pd.read_csv(os.path.join(os.path.dirname(notebook_path), "Data\\meta-data.csv"),na_values=" ")
# Getting Score test
    # Laterality Quotient. 
laterality = (metadat[['EHQ1',
                 'EHQ2',
                 'EHQ3',
                 'EHQ4',
                 'EHQ5',
                 'EHQ6',
                 'EHQ7',
                 'EHQ8',
                 'EHQ9',
                 'EHQ10']]-3)*25 
metadat["laterality"] =laterality.sum(axis=1)
metadat = metadat.drop(22)

# Adding a Histogram with the Age Distribution and Pie Chart With Sex 
sigma = np.std(metadat.AGE)
mu = np.mean(metadat.AGE)
explode = (0.1,0)
SEX, allvals = np.unique(metadat.SEX, return_counts=True)
#colors
colors = ['#26C281','#F5D76E']

def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d} )".format(pct, absolute)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))
fig.suptitle('Participants Data', fontsize=18 ,x = 0.2, y = 1.01)

ax1.pie(allvals,explode=explode, colors=colors,autopct=lambda pct: func(pct, allvals),
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.legend( ['Femenine','Masculine'],
          title="Gender",
          loc="center left",
          bbox_to_anchor=(0.1, 0, 0.5, 1))

#ax1.plt.setp( ['Femenine','Masculine'], size=8, weight="bold")

ax1.set_title("(A) Gender Composition")

n, bins, patches = plt.hist(metadat.AGE, facecolor='#26C281', alpha=0.65)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title(' (B) Participants Age Histogram', loc='left')
plt.text(40,6, f'$\mu={mu:.1f},\ \sigma={sigma:.1f}$')
plt.axvline(mu, color = 'r', linestyle = 'dashed', linewidth = 2)
plt.grid(True)
plt.show()
```
